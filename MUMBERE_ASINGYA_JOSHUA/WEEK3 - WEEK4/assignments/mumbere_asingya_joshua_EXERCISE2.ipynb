```python
# Import libraries
import pandas as pd
```

```python
# Number 1
s1 = pd.Series([4, 65, 436, 3, 9])
s2 = pd.Series([7, 0, 3, 897, 9])

# Compare the elements of the two series
print(s1 == s2)
```

```python
# Number 2
series_a = pd.Series([2, 4, 6, 8, 14])
series_b = pd.Series([1, 3, 5, 7, 9])

result_add = series_a + series_b
result_sub = series_a - series_b
result_mul = series_a * series_b
result_div = series_a / series_b

print('Addition of series \n',result_add)
print('Subtraction of series \n',result_sub)
print('Multiplication of series \n',result_mul)
print('Division of series \n',result_div)
```
