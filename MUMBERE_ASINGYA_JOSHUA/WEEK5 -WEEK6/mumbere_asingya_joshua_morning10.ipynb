```python
# Working with ML Model Accuracy of different ML Model accuracy
# Create a model for Logistic Regression
# Helps solve problems based on classification, need to create an independent variable and dependent variable
```

```python
y = df.target.values
x_data = df.drop(['target'], axis = 1)
```

```python
# Normalize data | Split data for training and testing
# Formula: x = (X - X-Min)/(X-Max - X-Min)
x = ((x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)))
```

```python
# Split the data 80% training, 20% testing
X_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# Transpose matrices
X_train = X_train.T
y_train= y_train.T
x_test = x_test.T
y_test = y_test.T
```

```python
"""
# Jeff
    - Train/Test
    - Logistic Regression
    - Decision Tree
    - Confusion Matrix
    - Hierachical Clustering
    - Categorical Clustering
    - K-Means
    - Cross Validation
    - AUC -ROC Curves
    - KNN
"""
```

```python
# Train and Test
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Model Evaluation
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
